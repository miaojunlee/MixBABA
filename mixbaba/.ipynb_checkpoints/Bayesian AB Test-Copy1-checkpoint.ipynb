## Simulations

```python
np.random.seed(234)
result=[]

for x in range(2000):
    
    #initialize in the range of [100, 50000), this follows the “discrete uniform” distribution
    imps_ctrl=np.int(np.random.randint(low=100, high=50000, size=1)) 
    
    #initialize CR follows normal distribution with mean at 5% and standard deviation of 2%
    cr=np.random.normal(loc=0.05, scale=0.02)
    
    #conversion=clicks * simulated cr; if less than 0, use small number 
    if cr<0: cr=0.001
    convs_ctrl = np.int(np.floor(imps_ctrl*cr))
    
    #initialize the treatment clicks based on control clicks but with variation
    imps_test=np.int(imps_ctrl*(1+np.random.normal(loc=0, scale=0.06)))
    convs_test=np.int(convs_ctrl*(1+np.random.normal(loc=0, scale=0.03)))
    test_cr=convs_test/imps_test
    
    a_C, b_C = convs_ctrl+1, imps_ctrl-convs_ctrl+1
    beta_C = beta(a_C, b_C)
    
    a_T, b_T = convs_test+1, imps_test-convs_test+1
    beta_T = beta(a_T, b_T)

    #calculating the lift
    bayesian_lift=(beta_T.mean()-beta_C.mean())/beta_C.mean()
    prob=calc_prob_between(beta_T, beta_C)
    
    count = np.array([convs_ctrl, convs_test])
    nobs = np.array([imps_ctrl, imps_test])

    stat, pval = proportions_ztest(count, nobs)

    freq_lift = (convs_test/imps_test - convs_ctrl/imps_ctrl)/(convs_ctrl/imps_ctrl)
    
    result.append((imps_ctrl, convs_ctrl, cr, imps_test, convs_test, test_cr, 
                   bayesian_lift, prob, 
                   stat, pval, freq_lift))
```

## Random Sample

### Generate Test Data

```python
import scipy.stats as scs
from src.data import generate_data
import gc
```

```python
import scipy.stats as scs
import pandas as pd
# import numpy as np


def generate_data(N_A, N_B, p_A, p_B, days=None, control_label='A',
                  test_label='B'):
    """Returns a pandas dataframe with fake CTR data

    Example:

    Parameters:
        N_A (int): sample size for control group
        N_B (int): sample size for test group
            Note: final sample size may not match N_A provided because the
            group at each row is chosen at random (50/50).
        p_A (float): conversion rate; conversion rate of control group
        p_B (float): conversion rate; conversion rate of test group
        days (int): optional; if provided, a column for 'ts' will be included
            to divide the data in chunks of time
            Note: overflow data will be included in an extra day
        control_label (str)
        test_label (str)

    Returns:
        df (df)
    """

    # initiate empty container
    data = []

    # total amount of rows in the data
    N = N_A + N_B

    # distribute events based on proportion of group size
    group_bern = scs.bernoulli(N_A / (N_A + N_B))

    # initiate bernoulli distributions from which to randomly sample
    A_bern = scs.bernoulli(p_A)
    B_bern = scs.bernoulli(p_B)

    for idx in range(N):
        # initite empty row
        row = {}
        # for 'ts' column
        if days is not None:
            if type(days) == int:
                row['ts'] = idx // (N // days)
            else:
                raise ValueError("Provide an integer for the days parameter.")
        # assign group based on 50/50 probability
        row['group'] = group_bern.rvs()

        if row['group'] == 0:
            # assign conversion based on provided parameters
            row['converted'] = A_bern.rvs()
        else:
            row['converted'] = B_bern.rvs()
        # collect row into data container
        data.append(row)

    # convert data into pandas dataframe
    df = pd.DataFrame(data)

    # transform group labels of 0s and 1s to user-defined group labels
    df['group'] = df['group'].apply(
        lambda x: control_label if x == 0 else test_label)

    return df
```

### Simulation

```python
result=[]

for i in range(100):
#     np.random.seed(i)
    actual_control_cr=np.random.normal(loc=0.05, scale=0.02)
    actual_treatment_cr=actual_control_cr*(1+np.random.normal(loc=0, scale=0.06))
              
    print("True Control CR:", actual_control_cr) 
    print("True Treatment CR:", actual_treatment_cr)

    ab_data = generate_data(N_A=250000, 
                            N_B=250000, 
                            p_A=actual_control_cr, 
                            p_B=actual_treatment_cr, 
#                         days=7, 
                            control_label='C', 
                            test_label='T')
    
    print("done generating population round", i)

    for r in range(10):

        imps_ctrl=np.int(np.random.randint(low=100, high=50000, size=1)) 
        convs_ctrl=ab_data[ab_data['group']=='C'].sample(n=imps_ctrl)['converted'].sum()
        sample_control_cr=convs_ctrl/imps_ctrl

        imps_test=np.int(imps_ctrl*(1+np.random.normal(loc=0, scale=0.06)))
        convs_test=ab_data[ab_data['group']=='T'].sample(n=imps_test)['converted'].sum()
        sample_treatment_cr=convs_test/imps_test

        a_C, b_C = convs_ctrl+1, imps_ctrl-convs_ctrl+1
        beta_C = beta(a_C, b_C)
    
        a_T, b_T = convs_test+1, imps_test-convs_test+1
        beta_T = beta(a_T, b_T)

        #calculating the lift
        bayesian_lift=(beta_T.mean()-beta_C.mean())/beta_C.mean()
        prob=calc_prob_between(beta_T, beta_C)

        count = np.array([convs_ctrl, convs_test])
        nobs = np.array([imps_ctrl, imps_test])

        stat, pval = proportions_ztest(count, nobs)

        freq_lift = (convs_test/imps_test - convs_ctrl/imps_ctrl)/(convs_ctrl/imps_ctrl)
    
        result.append((i, actual_control_cr, actual_treatment_cr,
               r, 
               imps_ctrl, convs_ctrl, sample_control_cr, 
               imps_test, convs_test, sample_treatment_cr, 
               bayesian_lift, prob, 
               stat, pval, freq_lift))
        
        r+=1
        print("done with round:", r)
    
    del ab_data
    gc.collect()
```

```python
df = pd.DataFrame(result, columns=['Sample',
                                   'Actual_Control_CR', 'Actual_Treatment_CR',
                                   'Round', 
                                   'Control_Clicks', 'Control_Sale', 'Control_CR',
                                   'Treatment_Clicks', 'Treatment_Sale', 'Treatment_CR', 
                                   'Bayesian_Lift', 'Prob',
                                   'Z-value', 'P-value', 'Freq_Lift'])
```
